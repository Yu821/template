Plot correlation coefficient also.

```python
from scipy.stats import pearsonr

def corrfunc(x, y, ax=None, **kws):
    """Plot the correlation coefficient in the top left hand corner of a plot."""
    r, _ = pearsonr(x, y)
    # gca() means get current axes
    ax = ax or plt.gca()
    # Control where to plot in xy=()
    ax.annotate(f'ρ = {r:.2f}', xy=(.10, 1.05), xycoords=ax.transAxes)
```

### Multiple pie charts

```python
peng = sns.load_dataset("penguins")

sns.set(font_scale=1.2)

cols = ["sex", "species"]
nrow, ncol = 1, 2


fig1, axes = plt.subplots(nrow, ncol, figsize=(10,6))

for i in range(0, nrow):
    for j in range(0, ncol):
        if not cols: continue
        # axes[i][j] for more than one row
        ax = axes[j]
        
        col=cols.pop(0)
            
        count = peng.groupby(col)[col].count()
        labs = count.index.to_list()
        _ = ax.pie(count.values, labels=labs, autopct='%1.1f%%', startangle=90)
        _ = ax.set_title("Proportion of " + col)
        

    
```

```python
import matplotlib as mpl

# Control text size in the rectangle
mpl.rcParams["font.size"] = 13

from statsmodels.graphics.mosaicplot import mosaic

fig, ax = plt.subplots(1,1, figsize=(4,5))

col_list = ['#4878d0', '#ee854a']

# Get ratio Ex: ('Female', 'Adelie'): 21.0,
ratio_dct = peng.groupby(["sex","species"]).apply(lambda x: round(x.count()/peng.shape[0],2)*100)["sex"].to_dict()

# Rectagle properties. Font size cannot be set.
# https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Rectangle.html
props = lambda key: {'color': col_list[0] if 'Male' in key else col_list[1], "alpha":0.6}
# Label setting. Kis the tuple such as ('Female', 'Adelie')
labs = lambda k: k[1]+" "+k[0]+"\n"+str(ratio_dct[k])+"%"

_ = mosaic(data=peng, index=["sex", "species"], gap=0.02, properties=props, labelizer=labs, ax=ax)

plt.tick_params(axis='both', which='major', labelsize=13)
plt.title("Ratio between sex and species", fontsize=16, pad=10)
```

```python
import squarify    # pip install squarify (algorithm for treemap)


peng = sns.load_dataset("penguins")

r = peng.groupby("species")["body_mass_g"].mean()
r = pd.DataFrame(r)

# Colormap
cmap = matplotlib.cm.Blues

# colors setting
norm = matplotlib.colors.Normalize(vmin=r["body_mass_g"].min(), vmax=r["body_mass_g"].max())
c = [cmap(norm(value)) for value in r["body_mass_g"]]

squarify.plot(sizes=r["body_mass_g"], label=r.index, alpha=0.8, color=c)
plt.axis('off')
```
